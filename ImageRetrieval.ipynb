#UI Retrieval

```python
def ComputeIOU(boxA, boxB):
  xA = max(boxA[0], boxB[0])
  yA = max(boxA[1], boxB[1])
  xB = min(boxA[2], boxB[2])
  yB = min(boxA[3], boxB[3])

  interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)

  boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
  boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)

  iou = interArea / float(boxAArea + boxBArea - interArea)

  return iou
```

```python
from ultralyticsplus import YOLO, render_result
model_YOLO = YOLO('ultralyticsplus/yolov8s')
model_YOLO.overrides['conf'] = 0.25
model_YOLO.overrides['iou'] = 0.45
model_YOLO.overrides['agnostic_nms'] = False
model_YOLO.overrides['max_det'] = 1000
```

```python
from torch.utils.data import Dataset
from PIL import Image
import os

class RetrievalDataset(Dataset):
    def __init__(self, image_dir=None, vis_processors=None):
        self.image_dir = image_dir
        self.vis_processors = vis_processors
        self.image_files = os.listdir(image_dir)

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        image_path = os.path.join(self.image_dir, self.image_files[idx])

        raw_image = Image.open(image_path).convert("RGB")
        w, h = raw_image.size
        image = self.vis_processors["eval"](raw_image).unsqueeze(0).to(device)

        results = model_YOLO.predict(raw_image)

        people_index = (results[0].boxes.cls == 0).nonzero()
        bb_people = results[0].boxes.data.to(torch.long)[people_index][:,0,:].tolist()

        people_list = [[p[0] / w, p[1] / h, p[2] / w, p[3] / h] for p in bb_people]

        return {'image':image, 'image_path':image_path, 'people_list': people_list}
```

```python
image_dir = '/content/drive/MyDrive/datasets/Flickr_8k/Images'
data = RetrievalDataset(image_dir=image_dir, vis_processors=vis_processors)
```

```python
from usearch.index import Index
from PIL import Image

index = Index(ndim=256, metric="cos", dtype="f16")
num_results = 15
index.load("/content/drive/MyDrive/TruyVanThongTinDPT/DoAn/index.usearch")
ROI_coordinates = {
    'x_temp': 0,
    'y_temp': 0,
    'x_new': 0,
    'y_new': 0,
    'clicks': 0,
}
selected_section = pd.DataFrame({'x_start':[], 'y_start':[], 'x_end':[], 'y_end':[]})

def get_select_coordinates(img, evt: gr.SelectData):
    global selected_section
    sections = []
    # update new coordinates
    ROI_coordinates['clicks'] += 1
    ROI_coordinates['x_temp'] = ROI_coordinates['x_new']
    ROI_coordinates['y_temp'] = ROI_coordinates['y_new']
    ROI_coordinates['x_new'] = evt.index[0]
    ROI_coordinates['y_new'] = evt.index[1]
    # compare start end coordinates
    x_start = ROI_coordinates['x_new'] if (ROI_coordinates['x_new'] < ROI_coordinates['x_temp']) else ROI_coordinates['x_temp']
    y_start = ROI_coordinates['y_new'] if (ROI_coordinates['y_new'] < ROI_coordinates['y_temp']) else ROI_coordinates['y_temp']
    x_end = ROI_coordinates['x_new'] if (ROI_coordinates['x_new'] > ROI_coordinates['x_temp']) else ROI_coordinates['x_temp']
    y_end = ROI_coordinates['y_new'] if (ROI_coordinates['y_new'] > ROI_coordinates['y_temp']) else ROI_coordinates['y_temp']
    if ROI_coordinates['clicks'] % 2 == 0:
        sections.append(((x_start, y_start, x_end, y_end), "ROI of Face Detection"))
        selected_section = pd.DataFrame({'x_start':[x_start], 'y_start':[y_start], 'x_end':[x_end], 'y_end':[y_end]})
        return (img, sections)
    else:
        point_width = int(img.shape[0]*0.02)
        sections.append(((ROI_coordinates['x_new'], ROI_coordinates['y_new'],
                          ROI_coordinates['x_new'] + point_width, ROI_coordinates['y_new'] + point_width),
                        "Click second point for ROI"))
        return (img, sections)

def update_table():
  return selected_section

def _search_text(query):
    query_processed = text_processors["eval"](query)
    query_embedding = model.extract_features({'text_input': query_processed}, mode="text").text_embeds_proj[:,0,:]
    matches = index.search(query_embedding.cpu().numpy(), num_results)
    images = [Image.open(data[idx]['image_path']).convert("RGB") for idx in matches.keys]
    return images

def _search_text_draw(query):
    query_processed = text_processors["eval"](query)
    query_embedding = model.extract_features({'text_input': query_processed}, mode="text").text_embeds_proj[:,0,:]
    matches = index.search(query_embedding.cpu().numpy(), num_results)
    box = selected_section.iloc[-1].values.astype('int')
    box[0] /= 1920
    box[2] /= 1920
    box[1] /= 1120
    box[3] /= 1120
    iou_values = []
    for idx in matches.keys:
      max_iou = 0
      for person in data[idx]['people_list']:
        max_iou = max(max_iou, ComputeIOU(box, person))
      iou_values.append({'idx': idx, 'iou': max_iou})

    sorted_iou_values = sorted(iou_values, key=lambda x: x['iou'], reverse=True)

    sorted_indices = [item['idx'] for item in sorted_iou_values]

    images = [Image.open(data[idx]['image_path']).convert("RGB") for idx in sorted_indices]

    return images

def _search_image(query):
    query_processed = vis_processors["eval"](query).unsqueeze(0).to(device)
    query_embedding = model.extract_features({'image': query_processed}, mode="image").image_embeds_proj[:,0,:]
    matches = index.search(query_embedding.cpu().numpy(), num_results)
    images = [Image.open(data[idx]['image_path']).convert("RGB") for idx in matches.keys]
    return images

with gr.Blocks() as demo:

    with gr.Tab("Img2Img"):
      query = gr.Image(label='Image query', type="pil")
      outputs = gr.Gallery(rows=[3], columns=[5])
      submit = gr.Button(value='search')
      submit.click(_search_image, inputs=query, outputs=outputs)

    with gr.Tab("Text2Img"):
      query = gr.Textbox(lines=1, label='search query')
      outputs = gr.Gallery(rows=[3], columns=[5])
      submit = gr.Button(value='search')
      submit.click(_search_text, inputs=query, outputs=outputs)

    with gr.Tab("Text+draw 2 img"):
      with gr.Row():
          input_img = gr.Image(value='https://wallpapers.com/images/hd/blank-white-background-xbsfzsltjksfompa.jpg',label="Click")
          img_output = gr.AnnotatedImage(label="ROI",
                                              color_map={"ROI of Face Detection": "#9987FF","Click second point for ROI": "#f44336"})
      with gr.Row():
          table = gr.DataFrame(
            headers = ['x_start', 'y_start', 'x_end', 'y_end'],
            datatype = ["number", "number", "number", "number"]
          )
      with gr.Row():
          query = gr.Textbox(lines=1, label='search query')
      with gr.Row():
          submit = gr.Button(value='search')
      with gr.Row():
          outputs = gr.Gallery(rows=[3], columns=[5])

      submit.click(_search_text_draw, inputs=query, outputs=outputs)

      input_img.select(get_select_coordinates, input_img, img_output).then(update_table, None, table)

demo.launch(share=True)
```
